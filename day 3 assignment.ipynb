```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", data=fmri,kind="line", hue="event")
```

```python
sns.relplot(x="timepoint", y="signal", data=fmri,kind="line", hue="region")
```

```python
sns.relplot(x="timepoint", y="signal", data=fmri,kind="line", hue="subject")
```

```python
sns.catplot(x = 'region', y = 'timepoint', kind = 'box', data = fmri,hue= "event")
```

```python
sns.catplot(x = 'region', y = 'timepoint', kind = 'box', data = fmri,hue= "region")
```

```python
sns.catplot(x = 'region', y = 'timepoint', kind = 'box', data = fmri,hue= "subject")
```
